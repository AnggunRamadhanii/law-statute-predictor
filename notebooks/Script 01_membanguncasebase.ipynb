**Membangun Case Base**

```python
import os
import re
from google.colab import drive

drive.mount('/content/drive')
folder_asal = '/content/drive/MyDrive/SUBCPMK3PENALARAN/Penalaran_Komputer/data'
folder_tujuan = '/content/drive/MyDrive/SUBCPMK3PENALARAN/Penalaran_Komputer'

if not os.path.exists(folder_tujuan):
    os.makedirs(folder_tujuan)

files = sorted([f for f in os.listdir(folder_asal) if f.endswith('.txt')])

print(f"Ditemukan {len(files)} file untuk dibersihkan. Memulai proses cleaning...\n")

cleaned_texts = []

for index, filename in enumerate(files, start=1):
    file_path = os.path.join(folder_asal, filename)

    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        text = f.read()

    watermarks_kotor = [
        r'(?i)hkama\s+ahkamah\s+agung\s+repub',
        r'(?i)ahkamah\s+agung\s+republik\s+indonesia',
        r'(?i)mah\s+agung\s+republik\s+indonesia',
        r'(?i)mahkamah\s+agung\s+republik\s+indonesia',
        r'(?i)blik\s+indonesi[a]*',
        r'(?i)putusan\.mahkamahagung\.go\.[a-z]*',
        r'(?i)direktori\s+putusan',
        r'(?i)hkama'
    ]

    for kotoran in watermarks_kotor:
        text = re.sub(kotoran, '', text)

    text = re.sub(r'\n{3,}', '\n\n', text)
    text = re.sub(r'[ \t]+', ' ', text)

    cleaned_texts.append((index, text.strip()))

print("--- CLEANING SELESAI ---")
print("Data siap untuk disimpan ke Drive.")
```

```python
for index, text in cleaned_texts:
    nama_baru = f"case_{index:03d}.txt"
    output_path = os.path.join(folder_tujuan, nama_baru)

    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(text)

print("--- TAHAP SIMPAN SELESAI ---")
print(f"Semua file bersih sudah disimpan di: {folder_tujuan}")
```
